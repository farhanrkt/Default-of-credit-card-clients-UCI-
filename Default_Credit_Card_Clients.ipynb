```python
#create a heatmap to view correlations between features
plt.figure(figsize=(18, 15))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create a box plot
plt.figure(figsize=(18, 15))
sns.boxplot(data=df)
plt.title('Box Plot of df')
plt.show()
```

```python
#check for class imbalances
sns.histplot(df['default payment next month'], kde=True)
plt.title('Distribution of Target Variable')
plt.show()
```

```python
#print the exact count
df["default payment next month"].value_counts()
```

```python
#Visualize relationships of default with credit limit
sns.boxplot(x='default payment next month', y='LIMIT_BAL', data=df)
plt.title('Credit Limit vs. Default Status')
plt.show()
```
